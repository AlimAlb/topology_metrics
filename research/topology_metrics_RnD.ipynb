### Wrapper for metrics calculation


```python
import numpy as np
from typing import Dict
import numpy as np
import mtd
import rtd
import tensorflow as tf
from precision_recall import knn_precision_recall_features
from sklearn.metrics.pairwise import pairwise_kernels

#TODO: FIT и MMD добавить
class topology_metrics:
  def __init__(self) -> None:
    pass

  def mtd(self, P: np.ndarray, Q: np.ndarray, bs_1:int = 100, bs_2:int = 1000, n: int = 20, device:str = 'cpu') -> np.float32:
    return mtd.mtopdiv(P, Q, batch_size1=bs_1, batch_size2=bs_2, n = n, pdist_device=device)

  def rtd(self, P: np.ndarray, Q: np.ndarray, bs: int = 500, trials:int = 10, device = 'cpu') -> np.float32:
    return rtd.rtd(P, Q, batch = bs, trials = trials, pdist_device=device)

  def improved_pr(self, P: np.ndarray, Q: np.ndarray, nhood_sizes: list[int] = [1, 3, 10], row_batch_size: int = 256, col_batch_size: int = 512, num_gpus: int = 1) -> Dict[str, np.ndarray]:
    with tf.compat.v1.Session() as _:
      state = knn_precision_recall_features(
        P,
        Q,
        nhood_sizes=nhood_sizes, # за что отвечают эти nhoodsizes? что это?
        row_batch_size=row_batch_size, # за что отвечают батчсайзы тут?
        col_batch_size=col_batch_size,
        num_gpus=num_gpus,
      )
    return state


  def mmd_rbf(self, X, Y, gamma=1.0):
      Kxx = pairwise_kernels(X, X, metric='rbf', gamma=gamma)
      Kyy = pairwise_kernels(Y, Y, metric='rbf', gamma=gamma)
      Kxy = pairwise_kernels(X, Y, metric='rbf', gamma=gamma)

      n = len(X)
      m = len(Y)

      mmd2 = (Kxx.sum() - np.trace(Kxx)) / (n*(n-1)) \
         + (Kyy.sum() - np.trace(Kyy)) / (m*(m-1)) \
         - 2 * Kxy.mean()

      return np.sqrt(max(mmd2, 0))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_ring(n_points: int, dim: int, radius: float = 1.0, thickness: float = 0.0, center: np.ndarray | None = None, seed: int | None = None) -> np.ndarray:
    rng = np.random.default_rng(seed)
    points = rng.normal(size=(n_points, dim))

    norms = np.linalg.norm(points, axis=1, keepdims=True)
    random_directions = points / norms

    std = thickness / 4
    random_rads = rng.normal(loc=radius, scale=std, size=(n_points, 1))
    random_rads = np.maximum(random_rads, 0.0)

    thick_sphere_points = random_directions * random_rads

    if center is not None:
      thick_sphere_points += np.array(center)

    return thick_sphere_points

n_points = 500
dim = 2
radius = 1.0
thickness = 0.2
center = np.array([2.0, 1.0])
ring_points = generate_ring(n_points, dim, radius, thickness=thickness, center=center, seed=42)

plt.figure(figsize=(6, 6))
plt.scatter(ring_points[:, 0], ring_points[:, 1], s=10, alpha=0.7)
plt.title(f'2D ring, thickness {thickness},center {center}')
plt.grid(True)
```

Тут каким-то образом используется теорема о $P(r' < r) = V(r)/V(R) = (r/R)^{dim}$ <br/>
Пусть теперь P(r' < r) - равномерное распределние $U[0;1]$ <br/>
Тогда $U = (r/R)^{dim} => r = R*U^{1/dim}$

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sphere(n_points: int, dim: int, radius: float = 1.0, center: np.ndarray | None = None, seed: int | None = None) -> np.ndarray:
    rng = np.random.default_rng(seed)
    points = rng.normal(size=(n_points, dim))
    norms = np.linalg.norm(points, axis=1, keepdims=True)

    scale_factors = rng.uniform(0, 1, size=(n_points, 1))**(1/dim)
    scaled_points = (points / norms) * (radius * scale_factors)

    if center is not None:
        scaled_points += np.array(center)

    return scaled_points


n_points = 1000
dim = 2
radius = 1.0
center = np.array([0.5, -0.5])
points = generate_sphere(n_points, dim, radius, center=center, seed=42)

plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], s=10, alpha=0.7)
plt.title(f'2D sphere raduis {radius}, center {center}')
plt.grid(True)
```

```python
import numpy as np

def sample_mixture_four_gaussians(
    n_samples: int,
    mean: np.ndarray | list[float] = (0.0, 0.0),   # μ
    cov: np.ndarray | None = None,                 # Σ
    radius: float = 0.0,                           # расстояние между центрами (в whitened пространстве)
    seed: int | None = None,
) -> np.ndarray:
    """
    Генерирует точки из смеси 4 гауссиан как в статье:
    - глобальное распределение имеет те же mean и covariance, что и N(mean, cov)
    - компоненты всё дальше разъезжаются при увеличении radius,
      но общие моменты (μ, Σ) остаются теми же.

    Ограничение: размерность = 2 (как на картинке в статье).
    Параметр radius должен удовлетворять radius**2 < 2.
    """

    rng = np.random.default_rng(seed)

    mean = np.asarray(mean, dtype=float)
    assert mean.shape == (2,), "Функция работает только для 2D (mean.shape == (2,))"

    if cov is None:
        cov = np.eye(2)
    cov = np.asarray(cov, dtype=float)
    assert cov.shape == (2, 2), "cov должен быть матрицей 2x2"

    # Проверяем допустимость r
    if radius**2 >= 2.0:
        raise ValueError("radius^2 must be < 2, иначе компонентная дисперсия становится отрицательной")

    # Шаг 1. Выбеливаем: X = μ + L Z,  Z ~ N(0, I), где Σ = L L^T
    L = np.linalg.cholesky(cov)

    # Шаг 2. Строим смесь в whitened-пространстве (где эталон ~ N(0, I))
    # центры четырёх компонент
    centers = np.array([
        [ radius, 0.0],
        [-radius, 0.0],
        [ 0.0,  radius],
        [ 0.0, -radius],
    ])

    # ковариация каждой компоненты так, чтобы общая ковариация была I
    var_component = 1.0 - 0.5 * radius**2      # σ_c^2
    if var_component <= 0:
        raise ValueError("var_component <= 0, уменьшите radius")
    std_component = np.sqrt(var_component)

    # равные веса 1/4
    comp_idx = rng.integers(0, 4, size=n_samples)      # индексы компонент
    comp_means = centers[comp_idx]                     # центр для каждой точки
    z_whitened = comp_means + rng.normal(
        size=(n_samples, 2)
    ) * std_component                                  # точки в whitened-пространстве

    # Шаг 3. Возвращаемся к исходным μ, Σ:
    # X = μ + L Z, где Z имеет mean=0 и cov=I
    samples = mean + z_whitened @ L.T

    return samples
```

```python
import matplotlib.pyplot as plt

mean = np.array([0.0, 0.0])
cov = np.eye(2)

radii = [0.0, 0.5, 1.0, 1.3, 1.4, 1.41]   # до sqrt(2) ≈ 1.414
fig, axes = plt.subplots(1, len(radii), figsize=(4*len(radii), 4))

for ax, r in zip(axes, radii):
    pts = sample_mixture_four_gaussians(3000, mean=mean, cov=cov, radius=r, seed=42)
    ax.scatter(pts[:, 0], pts[:, 1], s=2, alpha=0.6)
    ax.set_title(f"r = {r}")
    ax.set_aspect("equal")
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def sample_mixture_n_gaussians(
    n_samples: int,
    mean: np.ndarray | list[float] = (0.0, 0.0),   # μ
    cov: np.ndarray | None = None,                 # Σ
    radius: float = 0.0,                           # расстояние до центров (в whitened-пространстве)
    n_components: int = 4,                         # число гауссиан в смеси
    seed: int | None = None,
) -> np.ndarray:
    """
    Генерирует точки из смеси n гауссиан в 2D так, чтобы
    глобальное распределение имело те же mean и covariance,
    что и N(mean, cov).

    Идея:
    - В whitened-пространстве базовое распределение ~ N(0, I).
    - Берём n компонент с центрами на окружности радиуса `radius`,
      равномерно по углу.
    - Дисперсия компоненты выбирается так, чтобы итоговая ковариация была I.
    - Затем возвращаемся в исходное пространство X = μ + L Z, где Σ = L L^T.

    Ограничения:
    - размерность = 2 (mean.shape == (2,))
    - cov — матрица 2x2
    - radius**2 < 2  (иначе дисперсия компоненты станет ≤ 0)
    """

    rng = np.random.default_rng(seed)

    mean = np.asarray(mean, dtype=float)
    assert mean.shape == (2,), "Функция работает только для 2D (mean.shape == (2,))"

    if cov is None:
        cov = np.eye(2)
    cov = np.asarray(cov, dtype=float)
    assert cov.shape == (2, 2), "cov должен быть матрицей 2x2"

    if n_components < 1:
        raise ValueError("n_components должен быть >= 1")

    # Проверяем допустимость radius
    # Для центров на окружности радиуса r в 2D справедливо:
    # E[μ_i μ_i^T] = (r^2 / 2) * I
    # Тогда Var(Z) = σ_c^2 I + (r^2 / 2) I = I  =>
    # σ_c^2 = 1 - r^2 / 2
    if radius**2 >= 2.0:
        raise ValueError("radius^2 must be < 2, иначе компонентная дисперсия становится отрицательной")

    # Шаг 1. Вычисляем L: Σ = L L^T
    L = np.linalg.cholesky(cov)

    # Шаг 2. Строим смесь в whitened-пространстве (где эталон ~ N(0, I))

    # центры n компонент на окружности радиуса `radius`
    angles = np.linspace(0.0, 2.0 * np.pi, n_components, endpoint=False)
    centers = np.stack(
        [radius * np.cos(angles), radius * np.sin(angles)],
        axis=1
    )  # shape (n_components, 2)

    # ковариация каждой компоненты так, чтобы общая ковариация была I
    var_component = 1.0 - 0.5 * radius**2      # σ_c^2
    if var_component <= 0:
        raise ValueError("var_component <= 0, уменьшите radius")
    std_component = np.sqrt(var_component)

    # равные веса 1 / n_components
    comp_idx = rng.integers(0, n_components, size=n_samples)  # индексы компонент
    comp_means = centers[comp_idx]                            # центр для каждой точки

    # точки в whitened-пространстве
    z_whitened = comp_means + rng.normal(
        size=(n_samples, 2)
    ) * std_component

    # Шаг 3. Возвращаемся к исходным μ, Σ:
    # X = μ + L Z, где Z имеет mean=0 и cov=I
    samples = mean + z_whitened @ L.T

    return samples
```

```python
import matplotlib.pyplot as plt

mean = np.array([0.0, 0.0])
cov = np.eye(2)

radii = [0.0, 0.5, 1.0, 1.3, 1.4, 1.41]   # до sqrt(2) ≈ 1.414
fig, axes = plt.subplots(1, len(radii), figsize=(4*len(radii), 4))

for ax, r in zip(axes, radii):
    pts = sample_mixture_n_gaussians(3000, mean=mean, cov=cov, radius=r, n_components= 10, seed=42)
    ax.scatter(pts[:, 0], pts[:, 1], s=2, alpha=0.6)
    ax.set_title(f"r = {r}")
    ax.set_aspect("equal")
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
